```python
import pandas as pd

BASE = "Dados/engajamento_tratado.xlsx"
df = pd.read_excel(BASE)
display(df.head())
```

```python
lista_tags = df["Tags"].unique().tolist()

print(lista_tags)
print("\n")
display(df["Tags"].value_counts())          # Tipos de tags e quantidade de ocorrências de cada uma
print(df["Tags"].isnull().sum())            # qtd de elementos nulos
print((~df["Tags"].isnull()).sum())         # qtd de elementos não nulos
```

```python
# UTILIZAÇÃO DO SPLIT E EXPLODE

df["Tags"] = df["Tags"].str.split("/")
df_novo = df.explode('Tags')
print(df_novo.Tags.unique())

display(df)
display(df_novo)
```

```python
# Fazendo apenas uma verificação
type(df.loc[0,"Tags"])
```

```python
resultado = df_novo.groupby(["Pessoas","Tags"])[["Curtidas","Comentários"]].mean().sort_values(by="Curtidas", ascending=False)
display(resultado.reset_index())
```

- Aqui podemos ver que quando envolvemos pessoas com promoções tivemos um resultado muito superior a qualquer outra combinação!
- No geral vemos o impacto de como pessoas nas publicações influenciam no engajamento


#### Agora iremos analisar comparações específicas dentro desse resultado:

```python
display(resultado.index)

for pessoas,tag in resultado.index.tolist():
    print(f"RESULTADO: {tag}")
    comparacao = resultado.loc[(slice(None),tag), : ] 
    if len(comparacao) == 2:
        comparacao.loc[("Percentual aumento")] = ((comparacao.loc[("S", tag)] / comparacao.loc[("N", tag)])-1) * 100
        display(comparacao.reset_index())
    else:
        display(comparacao)
```

#### <span style="color:yellow">Resultado:</span> 
- Nesse comparativo entre cada uma das tags com e sem pessoas, podemos ver que nos casos em que temos pessoas, tem um aumento de praticamente 100% nas curtidas, e um aumento de 200% nos comentaŕios em grande parte... Ou seja, pessoas = aumento enorme de engajamento
